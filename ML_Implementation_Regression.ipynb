```

import sklearn # machine learning 
import numpy as np # computation 
import pandas as pd # data management

# 1. load the dataset 

dataset = pd.read_csv("https://raw.githubusercontent.com/datasets/covid-19/main/data/key-countries-pivoted.csv")    # don't copy the direct link from dataset, go to "raw" and copy the link on the browser ! because when the file(dataset) is so big=> it won't show you the dataset , we need to go to "raw" and copy the link 
dataset.head(10)
# 2. to know the dimension  : how many rows , how many columns of the dataset 
dataset.shape   
# to know the overall understanding of dataset :
dataset.describe

# check the varible(Numberical (Quatitantive) / categorical(Quanlitative)) 
numerical_variable = dataset.columns[dataset.dtypes != 'object']
dataset[numerical_variable].head(5)

# categorical_variable = dataset.columns[dataset.dtypes == 'object']
# dataset[categorical_variable].head(5)




```

```
categorical_variable = dataset.columns[dataset.dtypes == 'object']
dataset[categorical_variable].head(5)
```

```
# See Missing values (Missing value : NaN)
dataset.isnull().head(5)  # to see the whole   
        # False : mean "no missing value" in that particular column
        # True : mean " there exist missing value " in that particular column
dataset[numerical_variable].isnull().head(5)
dataset[categorical_variable].isnull().head(5)

```

```
# Find out lenght of the whole dataset  
len(dataset)
```

```
# Find out the percentage of the missing values 
dataset.isnull().sum()/len(dataset)*100
  # percentage of the missing values for numerical variable 
dataset[numerical_variable].isnull().sum()/len(dataset)*100
```

```
# Replace 
dataset.fillna(1)   # if there exists the missing value in particular column => that particular column will be filled with value <1> .
```

```
dataset['Italy'].replace(0,1).head(5)
```

```
dataset.replace(3,"Piseth").head(5)
```

```
dataset['Italy'].replace([0,1],['KIT',168]).head(5)
# we can't replace more than one "variable" at the time . It shows an error . (Use different lines for different variable , is better )
```

```
# create rows and add it into the original dataset :
        # first , we must know the number of rows of our dataset : dataset.shape
        # values of variable , instructions about the variable 

row_inserted = pd.Series(['2022-05-20',168,12,1,2,'AAA',3,4,5], index=['Date','China','US','United_Kingdom	','Italy','France','Germany','Spain','Iran'])
dataset.append(row_inserted ,ignore_index = True)

```
